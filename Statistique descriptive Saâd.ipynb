```python
def process_and_plot(dataframe, plot_type='bar'):
    columns_to_exclude = ['moyenne_NB_VALD_hebdo_période_jo_2022_par_ligne',
                          'moyenne_NB_VALD_hebdo_période_jo_2022_par_ligne_par_direction',
                          'nombre_validations_totales_pour_la_ligne',
                          'Station']

    data = dataframe.drop(columns=columns_to_exclude)
    dataframe['Total'] = data.sum(axis=1)

    sum_by_station = dataframe.groupby('Station')['Total'].sum().sort_values(ascending=False)

    plt.figure(figsize=(10, 8))

    if plot_type == 'bar':
        sum_by_station.plot(kind='bar', color='skyblue')
        plt.xlabel('Station')
        plt.title('Fréquentation par station tout au long de la journée')
        plt.xticks(rotation=90)
    elif plot_type == 'pie':
        plt.pie(sum_by_station, labels=sum_by_station.index, autopct='%1.1f%%', startangle=140)
        plt.title('Répartition de la fréquentation par station')
    
    plt.tight_layout()
    plt.show()
```

#### 2nde étape : concaténation des tableaux et nettoyage des données

```python
# Création d'une liste contenant tous les DataFrames
mes_datas = [df1, df2, df3, df3bis, df4, df5, df6, df7, df7bis, df8, df9, df10, df11, df12, df13, df14]
```

```python
# Création d'un DataFrame pour stocker la fréquentation mensuelle par ligne de métro
resultat = pd.DataFrame(columns=['Lignes de métro', 'Fréquentation mensuelle'])

# Parcours des données
for i, df in enumerate(mes_datas, start=1):
    ligne = f'Ligne {i}'
    
    # Suppression des doublons de stations et calcul de la fréquentation mensuelle
    frequentation_mensuelle = (df.drop_duplicates(subset='Station')
                               .groupby('Station')['moyenne_NB_VALD_hebdo_période_jo_2022_par_ligne']
                               .sum().reset_index()['moyenne_NB_VALD_hebdo_période_jo_2022_par_ligne']
                               .sum() * 4)
    
    # Ajout des lignes df3bis et df7bis à df3 et df7 respectivement
    if i == 3:
        frequentation_mensuelle += sum(df['moyenne_NB_VALD_hebdo_période_jo_2022_par_ligne'].sum() * 4 for df in [df3, df3bis])
    elif i == 7:
        frequentation_mensuelle += sum(df['moyenne_NB_VALD_hebdo_période_jo_2022_par_ligne'].sum() * 4 for df in [df7, df7bis])
    
    # Ajout des résultats à un nouveau DataFrame
    resultat.loc[i-1] = [ligne, frequentation_mensuelle]

# Affichage du DataFrame résultat
print(resultat)
```

#### 3e étape : production d'un graphique 

```python
# Création du graphique à barres
plt.figure(figsize=(10, 6))
plt.bar(resultat['Lignes de métro'], resultat['Fréquentation mensuelle'], color='skyblue')
plt.xlabel('Lignes de métro')
plt.ylabel('Fréquentation mensuelle moyenne')
plt.title('Fréquentation mensuelle moyenne par ligne de métro pendant la période hors Jeux Olympiques')
plt.xticks(rotation=45)  # Rotation des étiquettes sur l'axe des x pour une meilleure lisibilité
plt.tight_layout()
plt.show()
```
