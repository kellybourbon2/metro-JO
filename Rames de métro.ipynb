```python
import pandas as pd

ligne=pd.read_excel(r'/workspaces/metro-JO/Données importées/type de rame selon ligne.xlsx')
capacité=pd.read_excel(r'Données importées/capacité selon type de rame.xlsx')
display(ligne)
display(capacité)
```

```python
capacité.loc[6, 'Type de rame'] = 'M77'
ligne.loc[8, 'Ligne'] = '7bis'

# Fusion des DataFrames "ligne" et "capacité" sur la colonne "Type de rame en 2023"
resultat = pd.merge(ligne, capacité, left_on='Type de rame en 2023', right_on='Type de rame', how='left')

display(resultat)
```
